### Q1

```python
from urllib.request import urlopen
import nltk
nltk.download('punkt')
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Mount_Everest'
response = requests.get(url)
data = BeautifulSoup(response.content, 'html.parser')
```

```python
text_content = data.get_text()
text_content
```

```python
word_tokens = nltk.word_tokenize(text_content)
```

```python
total_count = word_tokens.count('Mount')
total_count
```

```python
with open('mount.txt', 'w') as file:
    file.write(text_content.upper())
```

```python
title = data.title.string
title
```

```python
url = 'https://www.london.ac.uk/sites/default/files/study-guides/introduction-to-natural-language-processing.pdf'
response = requests.get(url)
```

```python
extracted_content = ""
```

```python
with open('temp.pdf', 'wb') as f:
    f.write(response.content)
```

```python
with pdfplumber.open('temp.pdf') as pdf:
    extracted_content = ""
    pages = [12, 13]  # Page numbers you want to extract (starting from 1)

    for page_num in pages:
        # pdfplumber page numbers are 0-indexed
        page = pdf.pages[page_num - 1]
        extracted_content += page.extract_text()
```

```python
print(extracted_content)
```

```python
with open('nlp.txt', 'w') as f:
    f.write(extracted_content)
```

```python
import requests
import docx
import nltk
```

```python
url = 'https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine-Learning-and-Natural-Language-Processing.docx'
response = requests.get(url)
```

```python
with open('temp.docx', 'wb') as f:
    f.write(response.content)
```

```python
document = docx.Document('temp.docx')
```

```python
paragraphs = document.paragraphs[:5]
for paragraph in paragraphs:
    print(paragraph.text)
```

```python
text_content = ""
for paragraph in document.paragraphs:
    text_content += paragraph.text
```

```python
word_tokens = nltk.word_tokenize(text_content)
total_count = word_tokens.count('The')
print(total_count)
```
