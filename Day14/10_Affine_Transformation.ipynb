```python
import cv2
import numpy as np

# load image
image = cv2.imread('../images/lenna.png')

# define transformation parameters
tx = 50 # translation along x-axis
ty = 200 # translation along y-axis
angle = -20 # rotation angle 
scale = 200/ 512 # scaling factor
shear_x = 0 # Shear along x-axis
shear_y = 0 # Shear along y-axis

# create transformation matrix
M = np.float32([[scale* np.cos(np.radians(angle)), -np.sin(np.radians(angle)), tx],
                [np.sin(np.radians(angle)), scale* np.cos(np.radians(angle)), ty]])

# apply transformation using warpAffine
img_affine = cv2.warpAffine(image, M, (512, 512))

# display original and affine transformed image
cv2.imshow('affine transformed', img_affine)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# load image
image = cv2.imread('../images/lenna.png')

# define transformation parameters
tx = 150 # translation along x-axis
ty = 200 # translation along y-axis
angle = 45 # rotation angle 
scale = 200 / 512 # scaling factor

# create transformation matrix
M = np.float32([[scale * np.cos(np.radians(angle)), -scale * np.sin(np.radians(angle)), tx],
                [scale * np.sin(np.radians(angle)), scale * np.cos(np.radians(angle)), ty]])

# apply transformation using warpAffine
img_affine = cv2.warpAffine(image, M, (512, 512))

# display original and affine transformed image
cv2.imshow('Original', image)
cv2.imshow('Affine Transformed', img_affine)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# load image
image = cv2.imread('../images/lenna.png')

# define transformation parameters
angle = 45 # rotation angle 
scale = 200 / 512 # scaling factor

# calculate center of the image for rotation
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

# create rotation matrix
M = cv2.getRotationMatrix2D(center, angle, scale)

# apply transformation using warpAffine
img_affine = cv2.warpAffine(image, M, (w, h))

# display original and affine transformed image
cv2.imshow('Original', image)
cv2.imshow('Affine Transformed', img_affine)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
