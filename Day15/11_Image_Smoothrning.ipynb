Averaging Kernel

```python
import cv2
import numpy as np

# Load an image
image = cv2.imread(r"..\images\lenna.png")

# Define a 5x5 averaging kernel
kernel = np.ones((5,5), np.float32) / 25

# Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(image, -1, kernel)

#Show the orugnal and smoothed image
cv2.imshow('Orignal', image)
cv2.imshow('Smoothed', smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image = np.hstack([image, smoothed_image])
```

```python
cv2.imshow('Both Images', new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Gaussian Kernel

```python
kernel_size = 7 # can be adjusted as per need
sigma = 0.35 * ((kernel_size- 1) * 0.5 -1) + 0.9 # Gaussian std. deviation
kernel = cv2.getGaussianKernel(kernel_size, sigma)
```

```python
sigma
```

```python
kernel
```

```python
# Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(image, -1, kernel)

#Show the orugnal and smoothed image
cv2.imshow('Orignal', image)
cv2.imshow('Smoothed', smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image = np.hstack([image, smoothed_image])
cv2.imshow('Both Images', new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Median Filter

```python
gray = cv2.imread('../images/lenna.png', 0)
```

```python
gray.shape
```

```python
import matplotlib.pyplot as plt
```

```python
kernel_size = 5
median = cv2.medianBlur(gray, kernel_size)
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(gray, cmap= 'gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap= 'gray'), plt.title('Median Filtered')
plt.show()
```

Blur only a selected portion (100, 300)

```python
import cv2
import matplotlib.pyplot as plt

# load the image
image = cv2.imread('../images/lenna.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image.shape
```

```python
height, width, channels = image.shape
```

```python
cropped = image[100:300, 100:300]
plt.imshow(cropped)
```

```python
cropped
```

```python
kernel_size = 15
median = cv2.medianBlur(cropped, kernel_size)
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(cropped, cmap= 'gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap= 'gray'), plt.title('Median Filtered')
plt.show()
```
