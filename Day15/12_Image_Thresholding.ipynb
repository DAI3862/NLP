Averaging Kernel

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Image
image = cv2.imread('../images/lenna.png', 0)

# define the threshold
binary_image = image > 127
```

```python
binary_image
```

```python
# show original and binary image
# Show original and binary image
plt.subplot(121)
plt.title('Original')
plt.imshow(image, cmap='gray')
plt.subplot(122)
plt.title('Inverse Binary')
plt.imshow(binary_image, cmap='gray')
```

```python
binary_image.dtype
```

```python
binary_image = binary_image.astype(np.uint8)* 255
```

```python
cv2.imshow('Binary Image', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Using cv2 functions

```python
ret, thresh1 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(131)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.subplot(132)
plt.title('Binary Image')
plt.imshow(thresh1, cmap='gray')
plt.subplot(133)
plt.title('Inverse Binary Image')
plt.imshow(thresh2, cmap='gray')
```
