```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../images/lenna.png', 0)
```

```python
# Sobel paramenters: ksize, dx, dy, ddepth
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3) # Sobel for Horizontal edges
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3) # Sobel for Vetical edges

# Combine hte Sobel images 
combined_sobel = cv2.add(sobelx, sobely)
```

```python
thresh, binary_sobel = cv2.threshold(combined_sobel, 120, 255, cv2.THRESH_BINARY)
```

```python

# Show original and binary image
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.subplot(142)
plt.title('SobelX Image')
plt.imshow(sobelx, cmap='gray')
plt.subplot(143)
plt.title('Sobely Image')
plt.imshow(sobely, cmap='gray')
plt.subplot(144)
plt.title('Final Image')
plt.imshow(binary_sobel, cmap='gray')
```

Canny Edge Detection

```python
import cv2

# Load the image
img = cv2.imread('../images/lenna.png', 0)

# Apply  Canny edge detection
edges = cv2.Canny(img, 50 ,150) # Adjust thresholds if needed

# Display results
cv2.imshow('Original', img)
cv2.imshow('Canny Edge', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python

# Show original and binary image
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(142)
plt.title('Sobel Image')
plt.imshow(binary_sobel, cmap='gray')
plt.subplot(143)
plt.title('Canny Image')
plt.imshow(edges, cmap='gray')
plt.subplot(144)
plt.title('Inv Canny Image')
plt.imshow(~edges, cmap='gray')
```
