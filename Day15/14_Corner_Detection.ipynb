```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
gray = cv2.imread('../images/circle.png', 0)
image = cv2.imread('../images/circle.png')

# conver image to float32
dst = np.float32(gray)

# apply Harris Corner Detection
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Result is dialated for making the corners, not important
dst = cv2.dilate(dst, None)

# Threshold for an optimal value, it may vary as per the Image
gray[dst > 0.01 * dst.max()] = [255]
image[dst > 0.01 * dst.max()] = [0, 0, 255]

# Display the result
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.title('Grayscale Image')
plt.imshow(gray, cmap='gray')
plt.subplot(122)
plt.title('Colored Image')
plt.imshow(image)
```
