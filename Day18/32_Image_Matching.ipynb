```python
import cv2
import matplotlib.pyplot as plt

# read images
img1 = cv2.imread('/content/eiphel1.png',0)
img2 = cv2.imread('/content/eiphel2.png',0)

figure, ax = plt.subplots(1,2, figsize=(16,10))
ax[0].imshow(img1, cmap='gray')
ax[1].imshow(img2, cmap='gray')
```

```python
# sift
sift = cv2.xfeatures2d.SIFT_create()
keypoint_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoint_2, descriptors_2 = sift.detectAndCompute(img2,None)
```

```python
len(keypoint_1), len(keypoint_2)
```

```python
# feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)


matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1,keypoint_1,img2,keypoint_2,matches[:50],img2,flags=2)
plt.imshow(img3), plt.show()
```
