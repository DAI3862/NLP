```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
```

```python
data = url.read()
```

```python
print(data)
```

```python
soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
soup.find('title')
```

```python
text = soup.get_text()
print(text)
```

```python
from nltk.tokenize import word_tokenize
from nltk import FreqDist
```

```python
token = word_tokenize(text)
freq = FreqDist(token)
freq.plot(20)
```

```python
from nltk.corpus import stopwords

swords = stopwords.words('english')
swords

no_sw = ' '.join([word.lower() for word in token if word.lower() not in swords and word.isalnum()])
no_sw
```

```python
freq_wo_stopwords = FreqDist(word_tokenize(no_sw))
freq_wo_stopwords.plot(10)
```
