Data Preparation

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
sent = 'Hello friends! How are you? I like Python Programming.'
```

```python
# 1. Tokenize
tokens1 = word_tokenize(sent)
tokens1
```

```python
# 2. Remove the punctuations
tokens2 = [token for token in tokens1 if token.isalnum()]
tokens2
```

```python
# 3. Remove stopwords
swords = stopwords.words('english')

tokens3 = [token for token in tokens2 if token.lower() not in swords]
tokens3
```

```python
# 4. Remove the suffixes from text
ps = PorterStemmer()
tokens4 = [ps.stem(token) for token in tokens3]
tokens4
```

```python
# Function for cleaning the data
def clean_text(text):
    ps = PorterStemmer()
    tokens1 = word_tokenize(text)
    tokens2 = [token for token in tokens1 if token.isalnum()]
    swords = stopwords.words('english')
    tokens3 = [token for token in tokens2 if token.lower() not in swords]
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4
```

TF*IDF Vectorizer

```python
# import the class
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# create the object
tfidf = TfidfVectorizer(analyzer=clean_text)
```

```python
# apply
x_new = tfidf.fit_transform(x)
x_new
```

```python
x_new.data
```

Cross_Validation

```python
# Import the function
from sklearn.model_selection import train_test_split
```

```python
# seperate the training and testing
X_train, X_test, y_train, y_test = train_test_split(x_new, y, 
                                                    random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

Build the machine learning classification model

```python
# Import the class
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, r2_score, log_loss, confusion_matrix, ConfusionMatrixDisplay
```

```python
# Create the object
svm = SVC(kernel='linear', random_state=0)
```

```python
import time
```

```python
%%time
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```
