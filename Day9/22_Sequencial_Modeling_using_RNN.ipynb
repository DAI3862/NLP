```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
int_to_char = dict(enumerate(alphabets))
```

```python
int_to_char
```

```python
char_to_int = dict((v, i) for i, v in enumerate(alphabets))
char_to_int
```

Data Preparation

```python
seq_length = 1
x = []
y = []
for i in range(len(alphabets) - seq_length):
  seq_in = [alphabets[i]]
  seq_out = alphabets[i + 1]
  x += [[char_to_int[seq_in[0]]]]
  y += [char_to_int[seq_out]]
  print(seq_in, '->', seq_out)
```

```python
x
```

```python
y
```

Prepare the Output Variable

```python
from keras.utils import to_categorical

y_new = to_categorical(y)
```

```python
y_new
```

Prepare input variables

```python
import numpy as np
```

```python
x = np.reshape(x, (25, 1, 1))
x
```

```python
x.shape
```

```python
x_new = x / 25 # z is part of data but not the input data,  so we have to consider z too. Therefore 25.
x_new
```

```python
x_new.shape
```

Build the Model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(1, 1), activation='relu'))
model.add(LSTM(32))
model.add(Dense(26, activation='softmax'))
```

Compile the model

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Train the model

```python
model.fit(x_new, y_new, epochs=1000, batch_size=1)
```

```python
model.evaluate(x_new, y_new, batch_size=1)
```

Predict on new data

```python
val = 'K'
val_int = char_to_int[val]
val_int = np.reshape(val_int, (1, 1, 1))
val_int = val_int / 25
val_int
```

```python
pred = model.predict(val_int)
pred
```

```python
int_to_char[pred.argmax()]
```
